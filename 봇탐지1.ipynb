```python
import requests
import re

def main():
    # 1) ì„¸ì…˜(Session) ê°ì²´ ìƒì„±
    # ì„¸ì…˜ì„ ì‚¬ìš©í•´ì•¼ ì¿ í‚¤ê°€ ìë™ìœ¼ë¡œ ì €ì¥ë˜ì–´ ë‹¤ìŒ ìš”ì²­ì— í•¨ê»˜ ì „ì†¡ë©ë‹ˆë‹¤
    s = requests.Session()

    # 2) ë¸Œë¼ìš°ì € ìœ ì‚¬í•œ User-Agent ì„¤ì • (User-Agent ê²€ì‚¬ë¥¼ ìš°íšŒí•˜ê¸° ìœ„í•¨)
    # Chrome, Firefox ë“± ë‹¤ì–‘í•œ User-Agent ì¤‘ í•˜ë‚˜ë¡œ ì§€ì •í•©ë‹ˆë‹¤
    s.headers.update({
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                      "(KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
    })

    # 3) /level3 í˜ì´ì§€ë¥¼ ë¨¼ì € ìš”ì²­í•˜ì—¬ ì„¸ì…˜ ì¿ í‚¤ì™€ HTML ë‚´ js_token íšë“
    url_level3 = "http://127.0.0.1:5000/level3"
    r = s.get(url_level3)
    if r.status_code != 200:
        print("level3 í˜ì´ì§€ ìš”ì²­ ì‹¤íŒ¨: ", r.status_code, r.text)
        return

    # HTMLì—ì„œ í† í° ì¶”ì¶œ (í…œí”Œë¦¿ ì˜ˆì‹œì— ë”°ë¼ ì •ê·œì‹ ë“±ì„ í™œìš©)
    # level3.html ë‚´ë¶€ì—ì„œ:  const token = "{{ session['js_token'] }}";
    # ì´ëŸ° ì‹ìœ¼ë¡œ ë“¤ì–´ìˆì„ ê°€ëŠ¥ì„±ì´ ë†’ìŠµë‹ˆë‹¤
    match = re.search(r'const\s+token\s*=\s*["\']([^"\']+)["\']', r.text)
    if not match:
        print("í† í°ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. HTML êµ¬ì¡°ë¥¼ í™•ì¸í•˜ì„¸ìš”.")
        return

    js_token = match.group(1)
    print("íšë“í•œ js_token:", js_token)

    # 4) /protected_api/articles ìš”ì²­
    # Referer í—¤ë” ì¶”ê°€ + GET íŒŒë¼ë¯¸í„°ì— token ì¶”ê°€
    url_articles = "http://127.0.0.1:5000/protected_api/articles"
    params = {
        "page": 1,
        "pageSize": 20,
        "token": js_token  # ìë°”ìŠ¤í¬ë¦½íŠ¸ ê²€ì¦(ì„¸ì…˜ í† í°)
    }
    # Refererì— level3 ë¬¸ìì—´ì´ ë°˜ë“œì‹œ ë“¤ì–´ê°€ë„ë¡ ì„¤ì •í•©ë‹ˆë‹¤
    headers = {
        "Referer": "http://127.0.0.1:5000/level3"
    }
    
    r2 = s.get(url_articles, params=params, headers=headers)

    if r2.status_code == 200:
        print("ì •ìƒ ì‘ë‹µ:", r2.json())  # JSON ë°ì´í„° ì¶œë ¥
    else:
        print("ì°¨ë‹¨ëœ ì‘ë‹µ:", r2.status_code, r2.text)

if __name__ == "__main__":
    main()
```
